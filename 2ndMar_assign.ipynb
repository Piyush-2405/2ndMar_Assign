Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that allows you to create a wide variety of high-quality plots and charts. It provides a convenient interface for creating static, animated, and interactive visualizations in Python.

Matplotlib is used for various purposes, including:

1.Exploratory Data Analysis (EDA): Matplotlib helps in visually exploring and understanding data, identifying patterns, and gaining insights.

2.Presentation and Reporting: Matplotlib enables the creation of visually appealing plots for presentations, reports, and publications.

3.Model Evaluation and Validation: It allows the visual representation of model outputs, such as comparing predicted and actual values or plotting model performance metrics.

4.Decision Making: Matplotlib helps in visualizing data to make informed decisions, by highlighting trends, comparisons, and relationships within the data.

5.Communication: Matplotlib aids in effectively communicating data-driven stories or results to a broader audience, making complex information more accessible and understandable.

Five common plots that can be created using the Pyplot module of Matplotlib are:

1.Line Plot: Displays data points connected by straight line segments, often used to represent trends or time series data.

2.Bar Plot: Represents data as rectangular bars, where the length of each bar corresponds to the value being plotted. It is useful for comparing categorical data.

3.Scatter Plot: Shows the relationship between two numerical variables by representing individual data points as dots on a two-dimensional plane. It helps identify patterns, clusters, or correlations.

4.Histogram: Visualizes the distribution of a single variable by dividing the data into intervals (bins) along the x-axis and displaying the frequency or count of data points in each bin.

5.Pie Chart: Represents data as slices of a circular pie, where each slice represents a category and the size of the slice corresponds to the proportion of that category in the whole.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot")
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots in a grid format, making it easier to compare and analyze different plots simultaneously. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x1,y1)
axs[0,0].set_title("Line 1")

axs[0,1].plot(x2,y2)
axs[0,1].set_title("Line 2")

axs[1,0].plot(x3,y3)
axs[1,0].set_title("Line 3")

axs[1,1].plot(x4,y4)
axs[1,1].set_title("Line 4")

plt.tight_layout()

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of plot that represents categorical data using rectangular bars. It displays the values of different categories as the heights or lengths of the bars, allowing for easy visual comparison between categories. Bar plots are commonly used to show and compare values across different categories or groups.

Bar plots are useful for various purposes, including:

1.Comparing Quantities: Bar plots allow for easy comparison of values between different categories or groups, making it visually clear which category has higher or lower values.

2.Categorical Analysis: Bar plots provide a convenient way to analyze and understand categorical data, such as comparing sales figures of different products or performance metrics of various teams.

3.Visualizing Rankings or Order: Bar plots can represent the ranking or order of different categories based on a specific metric, making it easy to identify the top or bottom performers.

4.Communicating Results: Bar plots are effective in presenting data-driven insights or findings to a wider audience. They simplify complex information and make it easier to understand and interpret.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(6,4))
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.title("Profit bar")

plt.figure(figsize=(6, 4)) 
plt.barh(company, profit)
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.title("Profit bar")

plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a type of plot that displays the summary statistics of a dataset. It provides a concise visual representation of the distribution of the data, including the median, quartiles, and any potential outliers. Box plots are particularly useful for comparing multiple datasets or identifying the presence of outliers in a dataset.

Here are some key features and uses of box plots:

Median: The line within the box represents the median value of the dataset, which is the value that divides the data into two equal halves.

Quartiles: The box represents the interquartile range (IQR), which spans from the first quartile (Q1) to the third quartile (Q3). The length of the box indicates the spread of the middle 50% of the data.

Whiskers: The whiskers extend from the box and indicate the range of the data. By default, the whiskers extend up to 1.5 times the IQR. Data points beyond the whiskers are considered outliers and are plotted individually.

Outliers: Individual data points that fall outside the whiskers are considered outliers. They are plotted as individual points or circles on the plot, helping to identify extreme values or potential data anomalies.

Box plots are used for various purposes, including:

Visualizing and comparing the distribution of multiple datasets or groups.
Identifying skewness, symmetry, or the presence of outliers in the data.
Providing a summary of the central tendency and spread of the dataset.
Assessing the spread of data across different categories or factors.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]  # Combine the datasets into a list

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Data")
plt.ylabel("Value")

plt.show()
```
